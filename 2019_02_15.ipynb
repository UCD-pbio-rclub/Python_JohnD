## Combining Data with Overlap

````python
a = pd.Series([np.nan, 2.5, 0.0, 3.5, 4.5, np.nan],index=['f', 'e', 'd', 'c', 'b', 'a'])
a
````

````python
b = pd.Series([0., np.nan, 2., np.nan, np.nan, 5.],index=['a', 'b', 'c', 'd', 'e', 'f'])
b
````

````python
np.where(pd.isnull(a), b, a)
#If a is null use b otherwise use a
````

````python
b.combine_first(a)
````

````python
df1 = pd.DataFrame({'a': [1., np.nan, 5., np.nan],
                    'b': [np.nan, 2., np.nan, 6.],
                    'c': range(2, 18, 4)})
df1
````

## Reshaping with Hierarchical Indexing

````python
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
                    index=pd.Index(['Ohio', 'Colorado'], name='state'),
                    columns=pd.Index(['one', 'two', 'three'],
                    name='number'))
data
````

````python
result = data.stack()
result
````

````python
result.unstack()
````

````python
result.unstack(0)
````

````python
result.unstack('state')
````

````python
data2 = pd.concat([s1, s2], keys=['one', 'two'])
data2
````

````python
data2.unstack()
````

## Pivoting “Long” to “Wide” Format

````python
data = pd.read_csv('examples/macrodata.csv')
data.head()
````

````python
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter, name='date')
periods
````

````python
columns = pd.Index(['realgdp', 'infl', 'unemp'], name='item')
columns
````

````python
data = data.reindex(columns=columns)
data.head()
````

````python
ldata = data.stack().reset_index().rename(columns={0: 'value'})
ldata.head()
````

````python
ldata[:10]
````

## Pivoting “Wide” to “Long” Format

````python
df = pd.DataFrame({'key': ['foo', 'bar', 'baz'],
                   'A': [1, 2, 3],
                   'B': [4, 5, 6],
                   'C': [7, 8, 9]})
df
````

````python
melted = pd.melt(df, ['key'])
melted
````

````python
reshaped = melted.pivot('key', 'variable', 'value')
reshaped
````

````python
reshaped.reset_index()
````

````python
pd.melt(df, id_vars=['key'], value_vars=['A', 'B'])
````

````python
pd.melt(df, value_vars=['A', 'B', 'C'])
````

````python
pd.melt(df, value_vars=['key', 'A', 'B'])
````

Please use hierarchical indexing to label another level of index to show which chromosome each gene on and give the index name 'chromosome'.
